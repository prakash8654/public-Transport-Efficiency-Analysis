```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
travel=pd.read_csv("/content/Dataset.csv")
print(travel)
print(travel.head)
display(travel.drop_duplicates())
def categorize_feedback(number_of_boardings):
    if number_of_boardings <= 4:
        return "good"
    elif number_of_boardings > 4 and number_of_boardings <= 6:
        return "average"
    elif number_of_boardings > 6:
        return "bad"

input_csv_file = '/content/Dataset.csv'
output_csv_file = '/content/preprocessed_dataset (1).csv'

with open(output_csv_file, 'w', newline='') as output_file:
    fieldnames = ["TripID", "StopName", "Feedback"]
    writer = csv.DictWriter(output_file, fieldnames=fieldnames)
    writer.writeheader()

    with open(input_csv_file, 'r', newline='') as input_file:
        reader = csv.DictReader(input_file)
        for row in reader:
            trip_id = row["TripID"]
            stop_name = row["StopName"]
            number_of_boardings = int(row["NumberOfBoardings"])
            feedback = categorize_feedback(number_of_boardings)

            writer.writerow({"TripID": trip_id, "StopName": stop_name, "Feedback": feedback})

print(f"Feedback data has been written to {output_csv_file}")

input_csv_file = '/content/preprocessed_dataset (1).csv'

punctual_trips = 0
total_trips = 0
with open(input_csv_file, 'r', newline='') as input_file:
    reader = csv.DictReader(input_file)
    for row in reader:
        feedback = row["Feedback"]

        if feedback in ["good", "average"]:
            punctual_trips += 1

        total_trips += 1
punctuality_rate = (punctual_trips / total_trips) * 100

print(f"Punctuality Rate: {punctuality_rate:.2f}% (Based on 'good' and 'average' feedback for {total_trips} trips)")

mean_boardings = travel['NumberOfBoardings'].mean()
median_boardings = travel['NumberOfBoardings'].median()
std_deviation = travel['NumberOfBoardings'].std()

print(f"Mean boardings: {mean_boardings}")
print(f"Median boardings: {median_boardings}")
print(f"Standard Deviation: {std_deviation}")

route_avg_boardings = travel.groupby('RouteID')['NumberOfBoardings'].mean()
print("\nRoute-Level Analysis:")
print(route_avg_boardings)

stop_avg_boardings = travel.groupby('StopName')['NumberOfBoardings'].mean()
print("\nStop-Level Analysis:")
print(stop_avg_boardings)

week_avg_boardings = travel.groupby('WeekBeginning')['NumberOfBoardings'].mean()
print("\nTemporal Analysis:")
print(week_avg_boardings)
df.to_csv('analysis.csv', index=False)

```
